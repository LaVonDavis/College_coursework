```python
from gensim.models.word2vec import Word2Vec
import gensim.downloader as api
```

```python
# download the corpus and return it opened as an iterable
corpus = api.load('text8')
model = Word2Vec(corpus) # train a model from the corpus
print("Number of Vocab:",len(model.wv.key_to_index.keys()))
print("Feature Dimension:", model.vector_size)
print(model.wv.most_similar("car"))
```

```python
model.wv.save_word2vec_format("my_w2v_model.bin", binary=True)
print("Model save successfully!")
```

```python
sentences = [['i', 'like', 'apple', 'pie', 'for', 'dessert'],     ['i', 'dont', 'drive', 'fast', 'cars'],     ['data', 'science', 'is', 'fun'],     ['chocolate', 'is', 'my', 'favorite'],     ['my', 'favorite', 'movie', 'is', 'predator']]
```

```python
w2v = Word2Vec(sentences, min_count=1, vector_size = 5)
print("Model's Hyperparameters")
print(w2v)
# access vector for one word
print(w2v.wv.get_vector ('chocolate'))
```

```python
#list the vocabulary words
print("Word2Vec Vocabulary")
print(list(w2v.wv.key_to_index.keys()))
```

```python
w2v.wv.save_word2vec_format("my_w2v_model2.bin", binary=True)
#w2v.save("my_w2v_model.modal")
print("Model save successfully!")
```
