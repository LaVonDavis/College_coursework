```python
from gensim.models.word2vec import KeyedVectors
```

```python
pretrainedpath = "/content/drive/MyDrive/Colab Notebooks/GoogleNews-vectors-negative300.bin.gz"
```

```python
w2v_model = KeyedVectors.load_word2vec_format(pretrainedpath, binary=True)
print('done loading Word2Vec')
```

```python
print(len(list(w2v_model.key_to_index.keys())))
print(w2v_model.most_similar('beautiful'))
print(w2v_model['beautiful'])
```
