```python
import nltk
import pandas as pd
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#read data into dataframe
df = pd.read_csv('unlabeled_dataset.csv')
print(df.head())
```

```python
review_text = df['Review']
review_title = df['Title']

for title, review in zip(review_title, review_text):
  # Convert title and review to string and handle NaN values
  title = str(title)
  review = str(review)

  # Check if title or review are NaN after conversion
  if title == 'nan':
      score1 = {'compound': 0}  # Assign default score for NaN titles
  else:
      score1 = SentimentIntensityAnalyzer().polarity_scores(title)

  if review == 'nan':
      score2 = {'compound': 0}  # Assign default score for NaN reviews
  else:
      score2 = SentimentIntensityAnalyzer().polarity_scores(review)


  if score1['compound'] > 0:
      score1 = 1
  else:
      score1 = 0
  if score2['compound'] > 0:
      score2 = 1
  else:
      score2 = 0
  df.loc[df['Title'] == title, 'T_label'] = int(score1)  # Assign label based on title
  df.loc[df['Review'] == review, 'R_label'] = int(score2)  # Assign label based on review

df.to_csv('labeled_dataset.csv', index=False)
df = pd.read_csv('labeled_dataset.csv') # Fix: reassign df to the newly read CSV
print(df)
```
