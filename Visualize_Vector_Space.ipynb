```python
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from gensim.models.word2vec import KeyedVectors, Word2Vec
```

```python
#fname_path = "my_w2v_model.modal"
fname_path = "my_w2v_model2.bin"
```

```python
#load w2vec model
w2v = KeyedVectors.load_word2vec_format(fname_path, binary=True)
# w2v = Word2Vec.load(fname_path,)
# get model feature vectors
# X = [ w2v.wv.get_vector(v) for v in w2v.wv.key_to_index.keys()]
X = [ w2v.get_vector(v) for v in w2v.key_to_index.keys()]

#create pca (dimensionality reduction) object
pca = PCA(n_components=2)

#transform to 5D to 2D
result = pca.fit_transform(X)
```

```python
# create a scatter plot of the projection
plt.scatter(result[:, 0], result[:, 1])
```

```python
# words = list(w2v.wv.key_to_index.keys())
words = list(w2v.key_to_index.keys())
for i, word in enumerate(words):
  plt.annotate(word, xy=(result[i, 0], result[i, 1]))

plt.show()
```
