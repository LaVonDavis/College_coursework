```python
vocab={'dog':0,'bites':1,'man':2, 'meat':3, 'food':4,'eats':5}
processed_corpus=['dog bites man',
'dog eats meat',
'man eats food' ]
corpus_vector=[]
def get_onehot_vector(doc_string):
  onehot_encoded = []
  for word in doc_string.split():
    temp = [0]*len(vocab)
    if word in vocab:
      temp[vocab[word]] = 1
    onehot_encoded.append(temp)
  return onehot_encoded

for doc in processed_corpus:
  corpus_vector.append(get_onehot_vector(doc))
print(corpus_vector)
```
