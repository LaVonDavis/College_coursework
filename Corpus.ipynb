```python
import string
```

```python
corpus = ['python is fun', 'python is not only fun but also powerful', 'learning python is fun']
```

```python
def preprocess(text:str):
    #convert text to lowercase
    text = text.lower()
    #remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    #tokenize the text
    token = text.split()
    return token
```

```python
processed_corpus = [preprocess(sentence) for sentence in corpus]
print(processed_corpus)
```

```python
#initialize an empty set for corpus' vocab
vocab = set()
#build vocab
for sentence in processed_corpus:
  vocab.update(sentence)

vocab = sorted(list(vocab))
print(vocab)
```

```python
#calculate word fequency in each document in corpus
def create_bow_freq(sentence, vocab:list):
  word_freq = [0] * len(vocab)
  for word in sentence:
    if word in vocab:
      idx = vocab.index(word) #finds index of wordin vocab
      word_freq[idx] += 1 #increments count at that index
  return word_freq
```

```python
#create bow(bag of words) vector for each sentence in the processed corpus
bow_vectors = [create_bow_freq(sentence, vocab) for sentence in processed_corpus]
print(bow_vectors)
```
