```python
#Library
import spacy
import spacy.displacy
```

```python
#load and create model object
nlp =spacy.load("en_core_web_md")
# define the named entity categories that we want to recognize
ner_categories=["PERSON", "ORG", "GPE","PERSON"]
```

```python
#tokenize the text document
doc ="Robert Downey Jr. is an American actor and producer. " + \
  "He is best known for his roles in films such as Iron Man," + \
    "The Avengers, and Sherlock Holmes. Downey has won several awards " + \
      "for his acting, including two Screen Actors Guild Awards and a " + \
        "Golden Award. He has also nominated for an Academy Award."
doc = nlp(doc)
```

```python
#identify and classify name entities
entities =[]
for ent in doc.ents:
  if ent.label_ in ner_categories:
    entities.append((ent.text, ent.label_))
#visualize the enitities in the document
spacy.displacy.render(doc, style="ent")
```

```python
# Define a simple knowledge base (replace with your actual KB)
kb = {  "Barack Obama": {"type": "PERSON", "description": "44th President of the United States"},  "United States": {"type": "GPE", "description": "A country in North America"},  "Apple": {"type": "ORG", "description": "Technology company"},  "Google": {"type": "ORG", "description": "Technology company"},  "New York": {"type": "GPE", "description": "City in the US"}, }
```

```python
# Function to disambiguate and link entities
def disambiguate_and_link(text, knowledge_base):
  doc = nlp(text)
  linked_entities = {}
  for ent in doc.ents:
    entity_text = ent.text
    entity_type = ent.label_
    # Check if entity is in knowledge base
    if entity_text in knowledge_base:
      linked_entities[entity_text] = knowledge_base[entity_text]
    else:
      # Handle cases where entity is not in KB
      print(f"Entity '{entity_text}' not found in knowledge base.")
  return linked_entities
```

```python
# Example usage
text = "Barack Obama was the 44th president of the United States. Apple and Google are based in New York."
linked_entities = disambiguate_and_link(text, kb)

# Print the results
print("Original Text:", text)
print("\nLinked Entities:")

for entity, details in linked_entities.items():
  print(f"- {entity}: {details['type']} ({details['description']})")
```
